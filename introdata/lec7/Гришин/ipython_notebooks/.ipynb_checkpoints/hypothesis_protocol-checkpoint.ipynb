### Словарь проверки статистических гипотез

* **Понятие ошибки I и II рода**

| Действие/истина  | $H_0$  верна        | $H_1$  верна       |
| -----------------|:-------------------:|:------------------:|
| отклонить $H_0$  | Ошибка первого рода | Верное решение     |
| принять $H_0$    | Верное решение      | Ошибка второго рода|

* ** Ошибка первого рода**

Ошибка первого рода - отклонение верной нулевой гипотезы;

* ** Ошибка второго рода**

Ошибка второго рода - принятие ложной нулевой гипотезы;

* **Уровень значимости**

Уровень значимости - вероятность допустить ошибку первого рода;

* **Взаимосвязь ошибки I и II рода**

Если размер выборки фиксирован, то уменьшение уровня значимости будет вести к увеличению вероятности ошибки второго рода. Если ставится цель одновременного уменьшения обоих ошибок, то необходимо увеличить объем выборки.

* **Мощность критерия**

$$ Power(t) = 1 - P(Type\,\,II\,\,error), $$
то есть, мощность критерия - это вероятность отвергнуть нулевую гипотезу, в то время как верна альтернативная (не допустить ошибку II рода).

* **Виды статистических тестов**

| Наименование     | Вид альтернативной гипотезы| 
| -----------------|:--------------------------:|
| Левосторонний    | $H_1: \theta < \theta_0  $ |
| Правосторонний   | $H_1: \theta > \theta_0  $ |
| Двусторонний     | $H_1: \theta \ne \theta_0  $ |

* **Критическая область**
Регион области значения тестовой статистики, попадание в который ведет к отклонению нулевой гипотезы;

* **p-value**
Ответ на вопрос: "Если верна нулевая гипотеза, какова вероятность наблюдать значение тестовой статистики большее, нежеле рассчитанное?"

* **Интерпретация значений p-values**

| p-value               | Интерпретация (для $\alpha = 0.05$)| 
| ----------------------|:----------------------------------:|
| $p <0.01$             |Очень сильное свидетельство против $H_0$  |
| $0.01-0.05$           |Сильное свидетельство против $H_0$  |
| $0.05-0.10$           |Слабое свидетельство против $H_0$  |
| $p >0.1$              |Нет свидетельства против $H_0$|

### Общий протокол проверки статистических гипотез

1. Сформулировать нулевую $H_0$ и альтернативную $H_1$ гипотезы.
2. Сформулировать статистические допущения о распределении выборки, в частности зависимости между наблюдениями в модели.
3. В зависимости от статистичсеких допущений, принятых на шаге 2, следует выбрать тестовую статистику, отвечающую характеру предположений.
4. Для тестовой статистики следует определить распределении при условии верности $H_0$.
5. Следует определить уровень значимости $\alpha$, который выступит пороговым значением для принятие отклонения гипотезы.
6. Рассчитать наблюдаемое значения тестовой статистики - $t_{obs} = f(X_1, ..., X_n)$.
7.  

   * **Подход на основе критических регионов**:
   Если наблюдаемое значение попадает в критический регион (например, не входит в $1-\alpha$ доверительный регион теоретических значений тестовой статистики при условии верной $H_0$), то нулевую гипотезу следует отклонить;
   
   
   * **Подход на основе p-values**:
   Посчитать p-value - вероятность наблюдать $T > t_{obs}$ если нулевая гипотеза верна. Полученное значение сравнить с уровнем значимости. Если p-value меньше или равно уровню значимости, то нулевую гипотезу следует отклонить.


### Параметрические тесты

Класс тестов, который предполагают, что наблюдаемая выборка относится к такому типу распределения, который может быть описан с помощью конечного числа параметров.

### Параметрические тесты для одной выборки

* **1-sample z-test**

Проверка гипотезы о величине истинного значения параметра генеральной совокупности $\mu$. Предполагает знание 2 параметра - генеральной дисперсии либо очень большой выборки.

* **1-sample t-test**

Проверка гипотезы о величине истинного значения параметра генеральной совокупности $\mu$. Используется, когда известна только оценка стандартного отклонения или на маленьких выборках (вспомните историю про Стьюдента!).

* **1-sample proportion test**

Проверка гипотезы о величине истинного значения доли в генереральной совокупности.

* **1-sample variance test**
Проверка гипотезы о величине дисперсии в генеральной совокупности.

* **Тест Шапиро-Уилка**

Данный тест используется для проверки нулевой гипотезы о номральном распределении генеральной совокупности.

* **Хи-квадрат Пирсона**

Данный тест используется для проверки нулевой гипотезы о том, что генеральная совкупность подчиняется некоторому закону распределения $F(x)$.

### Параметрические тесты для нескольких выборок

* **t-test для средних**

Имеет несколько вариантов в зависимости от того, какая информация нам доступна и допущениях о независимости выборок.

* **z-test для доли**

Используется для проверки гипотезы о равенстве доли в двух совокупностях. Использует нормальное распределение для аппроксимации выборочной статистики.

* **t-test для коэффициента корреляции Пирсона**

Используется для проверки гипотезы о том, что линейная связь между двумя признаками на самом деле отсутствует.
