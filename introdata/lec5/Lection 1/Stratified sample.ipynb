### Стратифицированная и нестратифицированная выборки

```python
%matplotlib inline
import numpy as np
from sklearn import cross_validation
import pandas as pd
import matplotlib.pyplot as plt
```

Допустим, у нас есть население, состоящее из бедных и богатых людей, в кажой группе есть некоторая вариативность доходов:

```python
poor = np.random.uniform(low = 10, high = 100, size = 1000)
rich = np.random.uniform(low = 1000, high = 2000, size = 20)

poor_label = np.repeat("poor", 1000)
rich_label = np.repeat("rich", 20)
```

Богатых мало, но обладают значительной долей дохода:

```python
np.sum(poor)
```

```python
np.sum(rich)
```

```python
np.sum(poor)+np.sum(rich)
```

```python
population = np.append(poor, rich)
pop_labels = np.append(poor_label, rich_label)
```

Представим, что мы хотим определить суммарный доход населения по выборке.

Симулируем стратифицированную и нестратифицированную выборки:

```python
non_stratified_estimate = []
attempts = 100
for j in range(attempts):
    sample = np.random.choice(population, size = 102, replace = False)
    non_stratified_estimate.append(np.sum(sample)*10)
```

```python
stratified_estimate = []
attempts = 100
for j in range(attempts):
    sample, nonsample = cross_validation.train_test_split(population, train_size = 102, stratify = pop_labels)
    stratified_estimate.append(np.sum(sample)*10)
```

Оценим результат на графике:

```python
scatter = np.random.normal(0,1,100)

plt.plot(scatter, non_stratified_estimate , 'o', label = 'non_stratified_estimate')
plt.plot(scatter, stratified_estimate , 'o', label = 'stratified_estimate')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2,)
```
