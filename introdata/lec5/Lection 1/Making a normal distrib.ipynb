### Как получается нормальное распределение

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

Будем собирать нормальное распределение из равномерных:

```python
sim_size = 10000000

first_uniform = np.random.uniform(0, 1, sim_size)
second_uniform = np.random.uniform(0, 1, sim_size)
third_uniform = np.random.uniform(0, 1, sim_size)



def standardize(x):
    return (x - np.mean(x)) / np.std(x)

## create normal distribution
normal_dist = np.random.normal(0,1, sim_size)
normal_dist1 = np.random.normal(0,1, sim_size)

binseq = np.linspace(-5, 5, num= 1001)


def overlap(dist1, dist2):
    
    hist1, edge = np.histogram(dist1, bins = binseq)
    hist2, edge = np.histogram(dist2, bins = binseq)
    
    diff = (2.0*sim_size - np.sum(np.abs(hist1 - hist2)))/(2*sim_size)
    
    return diff

total_1 = standardize(first_uniform)
total_2 = standardize(first_uniform + second_uniform)
total_3 = standardize(first_uniform + second_uniform + third_uniform)

x = np.linspace(-3, 3, 100)
```

#### Оценим ошибку метрики:

```python
overlap(normal_dist, normal_dist1)
```

#### 1 равномерное распределение:

```python
fig, (ax1) = plt.subplots(1,1, figsize=(5, 5))

ax1.plot(x, stats.norm.pdf(x, loc = 0, scale = 1), "r")
ax1.hist(total_1, bins = 100, normed = True)[2]
```

```python
overlap(normal_dist, total_1)
```

#### 2 равномерных распределения:

```python
fig, (ax1) = plt.subplots(1,1, figsize=(5, 5))

ax1.plot(x, stats.norm.pdf(x, loc = 0, scale = 1), "r")
ax1.hist(total_2, bins = 100, normed = True)[2]
```

```python
overlap(normal_dist, total_2)
```

#### 3 равномерных распределения:

```python
fig, (ax1) = plt.subplots(1,1, figsize=(5, 5))

ax1.plot(x, stats.norm.pdf(x, loc = 0, scale = 1), "r")
ax1.hist(total_3, bins = 100, normed = True)[2]
```

```python
overlap(normal_dist, total_3)
```
