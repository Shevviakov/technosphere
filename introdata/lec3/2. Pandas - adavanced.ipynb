# 2.2. groupby()

```python
df = pd.read_csv('beer_subset.csv.gz', parse_dates=['time'], compression='gzip')
review_cols = ['review_appearance', 'review_aroma', 'review_overall',
               'review_palate', 'review_taste']
df.head()
```

```python
df.beer_id.nunique()
```

```python
df.profile_name.nunique()
```

** groupby() ** - является композицей следующих:

* разбиение исходной таблицы на группы;
* применение функции в каждой группе;
* объединение результатов.

```python
gr = df.groupby('beer_style')
```

```python
gr
```

```python
gr.agg('mean')
```

Можно индексировать объекты дл группировки по маске:

```python
gr[review_cols].agg('mean')
```

Нечисленные стобцы исключаются.

```python
gr.abv.agg('mean')
```

```python
gr.abv.mean()
```

Можно к одной группировке применять сразу аггрегирующих функций:

```python
gr['review_aroma'].agg([np.mean, np.std, 'count']).head()
```

```python
gr[review_cols].agg(['mean', 'count', 'std'])
```

```python
gr[review_cols].agg(['mean', 'count', 'std']).stack(level=0)
```

Группировать можно и по нескольким столбцам:

```python
df.groupby(['brewer_id', 'beer_style']).review_overall.mean()
```

```python
df.groupby(['brewer_id', 'beer_style'])[review_cols].mean()
```

СЗ 2.1: 
* Найти автора, у которого разброс между лучшими и худшими оценками максимален;

```python
df.groupby('brewer_id')['review_overall'].agg('std').idxmax()
```

** melt **

Тяжело строго описать, что делает эта функция, проще показать "на пальцах" - из "широкого" формата переводит в "узкий": 

```python
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'}, 'B': {0: 1, 1: 3, 2: 5}, 'C': {0: 2, 1: 4, 2: 6}})
```

```python
df
```

```python
pd.melt(df, id_vars=['A'], value_vars=['B'], var_name='B_var', value_name='B_val')
```

Более интересный пример:

```python
column_names = ['date', '_', 'away_team', 'away_points', 'home_team',
                'home_points', 'n_ot', 'notes']
# обратите внимание на композицию методов ниже
games = (pd.read_csv('games.csv', names=column_names, parse_dates=['date'],
                     skiprows=1)
           .drop(['_', 'notes', 'n_ot'], axis='columns')
           .set_index('date', append=True))
games.index.names = ['game_id', 'date']
games.head()
```

```python
tidy = pd.melt(games.sort_index().reset_index(),
               id_vars=['game_id', 'date'], value_vars=['away_team', 'home_team'],
               value_name='team')
```

```python
tidy.head()
```

# Старайтесь приводить свои данные к следующему формату:

1. Каждая интересующая вас переменная - это столбец;
2. Каждое отдельное наблюдение  - это отдельная строка;
3. Дублирующих индексов быть не должно.

```python
tidy['rest'] = tidy.sort_values(by='date').groupby('team').date.diff().dt.days - 1
tidy.dropna().head()
```

```python
un = pd.pivot_table(tidy, values='rest',
                    index=['game_id', 'date'],
                    columns='variable').rename(
    columns={'away_team': 'away_rest', 'home_team': 'home_rest'}
)
un.columns.name = None
```

```python
un.dropna().head()
```

```python
df = pd.concat([games, un], axis=1)
df
```
